```python
def report_avg_gap(gt_mod,dt_mod):
    diff_sum=0.0
    norm = 0

    gt_mod = np.array(gt_mod)
    dt_mod = np.array(dt_mod)
    gt_mean = np.mean(gt_mod)
    dt_mean = np.mean(dt_mod)
    gt_max = np.max(gt_mod)
    dt_max = np.max(dt_mod)
    
    #for i in range(0, len(gt_mod)):
    #    gt_mod[i] = (gt_mod[i]*1.0)*1.0/gt_mean
    #    dt_mod[i] = (dt_mod[i]*1.0)*1.0/dt_mean
        
    #for i in range(0, len(gt_mod)):
    #    cur_diff = gt_mod[i] - dt_mod[i]
    #    if cur_diff >0:
    #        diff_sum = diff_sum + cur_diff
        #else:
        #    diff_sum = diff_sum - cur_diff
            
    for i in range(0,len(gt_mod)):
        cur_diff = ((gt_mod[i]-dt_mod[i])*1.0)/(gt_mod[i]*1.0)
        if cur_diff>0:
            diff_sum = diff_sum+cur_diff
           
            
    return (diff_sum*1.0)/len(gt_mod)
    
```

```python
alpha = [ 0.1, 0.2, 0.4,0.5, 0.6, 0.8,0.9]
pval =[ 0.6, 0.7,0.8,0.9]
```

```python
fsize = 18
```

```python
def alphaplots(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label):
    gt_mod = []
    mixmod_mod = []
    alllouvain = []
    
    for a in alpha:
        #print(a)
        netlist=[]
        gt_mod = []
        mixmod_mod = []
        alllouvain=[]
        titlestring = algoname+": Alpha="+str(a)
        for p in pval:
            strn='./Raphael_27.6.17/net_networks/network_' + str(a) + '_' + str(p) + '_0.05_0.3_0.0:'
            netlist.append(strn)

            resultsmixmod = open(resultsfiledir+resultsfilename)
            resultsmixmod.readline()
            for line in resultsmixmod:
                line = line.split()

                if(line[0] == strn):
                    #print("ppppppppp") 
                    gt_mod.append(float(line[1]))
                    mixmod_mod.append(float(line[2]))
                    #xval.append(p)
                    break
            if(compare==1):
                resultsalllouvain = open(comparewithfile)
                resultsalllouvain.readline()
                for line in resultsalllouvain:
                    line = line.split()

                    if(line[0] == strn):
                        alllouvain.append(float(line[2]))
                        #xval.append(p)
                        break
                resultsalllouvain.close()

            resultsmixmod.close()
            

        plt.figure(figsize = (10,10))
        plt.plot(pval,gt_mod, label = 'Ground Truth')
        plt.plot(pval,mixmod_mod, label = label)
        if(compare==1):
            plt.plot(pval,alllouvain,label = 'All_louvain')
        plt.legend(loc = 'lower right')
        #gap = report_avg_gap(gt_mod,mixmod_mod)
        titlestring = titlestring
        plt.title(titlestring,fontsize = fsize)
        plt.ylim([lowerlim, upperlim])
        plt.xlabel("P",fontsize = fsize)
        plt.ylabel("modularity",fontsize = fsize)
        #plt.show()
        plt.savefig(pathtosave+titlestring+'.png')
        plt.close();
```

```python
def pplots(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label):
    gt_mod = []
    mixmod_mod = []
    alllouvain = []
    
    for p in pval:
        #print(a)
        netlist=[]
        gt_mod = []
        mixmod_mod = []
        alllouvain=[]
        titlestring = algoname+":P="+str(p)
        for a in alpha:
            strn='./Raphael_27.6.17/net_networks/network_' + str(a) + '_' + str(p) + '_0.05_0.3_0.0:'
            netlist.append(strn)
            #print(strn)
            #print(netlist)
            resultsmixmod = open(resultsfiledir+resultsfilename)
            resultsmixmod.readline()
            for line in resultsmixmod:
                line = line.split()
                #print(line[0])

                if(line[0] == strn):
                    #print("ppppppppp") 
                    gt_mod.append(float(line[1]))
                    mixmod_mod.append(float(line[2]))
                    break

            if(compare):
                resultsalllouvain = open(comparewithfile)
                resultsalllouvain.readline()
                for line in resultsalllouvain:
                    line = line.split()

                    if(line[0] == strn):
                        alllouvain.append(float(line[2]))
                        #xval.append(p)
                        break
                resultsalllouvain.close()

            resultsmixmod.close()

        #print(len(gt_mod))
        plt.figure(figsize = (10,10))
        plt.plot(alpha,gt_mod, label = 'Ground Truth')
        plt.plot(alpha,mixmod_mod, label = label)
        if(compare):
            plt.plot(alpha,alllouvain,label = 'All_louvain')
        plt.legend(loc = 'lower right')
        #gap = report_avg_gap(gt_mod,mixmod_mod)
        titlestring = titlestring
        plt.title(titlestring,fontsize = fsize)
        plt.ylim([lowerlim, upperlim])
        plt.xlabel("alpha",fontsize = fsize)
        plt.ylabel("modularity",fontsize = fsize)
        #plt.show()
        plt.savefig(pathtosave+titlestring+'.png')
        plt.close();
```

```python
def plotalltogether(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label,saveas): 
    gt_mod = []
    mixmod_mod = []
    alllouvain = []
    titlestring = algoname
    axeslabels = []
    x = [] 
    i=1
    for a in alpha:
        #print(a)
        netlist=[]
        for p in pval:
            strn='./Raphael_27.6.17/net_networks/network_' + str(a) + '_' + str(p) + '_0.05_0.3_0.0:'
            netlist.append(strn)
            resultsmixmod = open(resultsfiledir+resultsfilename)
            resultsmixmod.readline()
            for line in resultsmixmod:
                line = line.split()

                if(line[0] == strn):
                    #print("line found")
                    #print("Line found alpha: {0} p: {1}".format(a, p))
                    gt_mod.append(float(line[1]))
                    mixmod_mod.append(float(line[2]))
                    axestick = str(a)+"/"+str(p)
                    axeslabels.append(axestick)
                    x.append(i)
                    i+=1
                    #xval.append(p)
                    break
            if(compare==1):
                resultsalllouvain = open(comparewithfile)
                resultsalllouvain.readline()
                for line in resultsalllouvain:
                    line = line.split()

                    if(line[0] == strn):
                        alllouvain.append(float(line[2]))
                        #xval.append(p)
                        break
                resultsalllouvain.close()

            resultsmixmod.close()
            

    plt.figure(figsize = (10,10))
    plt.plot(x,gt_mod, label = 'Ground Truth')
    plt.plot(x,mixmod_mod, label = label)
    if(compare==1):
        plt.plot(x,alllouvain,label = 'All_louvain')
    plt.legend(loc = 'lower right')
    #gap = report_avg_gap(gt_mod,mixmod_mod)
    titlestring = titlestring
    plt.title(titlestring,fontsize = fsize)
    plt.ylim([lowerlim, upperlim])
    plt.xlabel("Alpha/P values",fontsize = fsize)
    plt.ylabel("Modularity",fontsize = fsize)
    #print("x: ", x, "axeslabels: ",axeslabels)
    plt.xticks(x,axeslabels, rotation = 90)
    #plt.show()
    plt.savefig(pathtosave+saveas+'.png')
    plt.close();
```

```python
algoname = "MixMod_oldcode"
resultsfiledir = './resultsmixmod/modularity_comparisions/'
resultsfilename = 'modComparisionMixModLouvain_wt_oldcode_V2_with_cihi'
pathtosave = resultsfiledir+"alpha_p_plots_V2/oldcode/with_cihi/"
comparewithfile='./resultsOriginalModAlways/modComparisionOriginalLouvainAlways_newnets_oldmod_with_cihi'
label = "MixModLouvain_oldcodeV2"
compare=1
lowerlim=0.0
upperlim=1.0
alphaplots(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label)
pplots(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label)

pathtosave = resultsfiledir+"alpha_p_plots_V2/oldcode/"
saveas = "modComparision_newnets_oldmod_with_cihi"
plotalltogether(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label,saveas)
```
