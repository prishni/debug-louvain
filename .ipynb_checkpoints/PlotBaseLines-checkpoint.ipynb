```python
import sys,os
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
import pandas as pd
import seaborn as sns
from IPython.display import display
import copy
```

```python
folderpath = "./syntheticNetworkGeneration/netsForcomparingBaseline/results/"
final0 = folderpath +"mixmod_for_corrImpl_final0"
final1  =folderpath +"mixmod_for_corrImpl_final1"
union = folderpath + "union_fixing2"

allLouvainQ = folderpath +"all_louvain_for_corrImpl" 
mixmod_mQ = folderpath +"mixmod_mQ"             #This is final0 with mQ
grivan_Qm = folderpath + "grivanNewman_Qm"

infocom = folderpath +"nmi_infocom"

list_alpha = [0.2,0.4,0.6,0.8]
list_mu = [0.05,0.1,0.2,0.3]
list_p = [0.2,0.4,0.6,0.8]
list_p1 = [0.2,0.4,0.6,0.8]
list_p2 = [0.0,0.1,0.2,0.3]

plottypes = ['mod','nmi']
```

```python
def read_data():
    fileset1 = [final0,final1,allLouvainQ,mixmod_mQ,grivan_Qm,union]
    final0res =[]
    final1res=[]
    alllouvainres=[]
    mixmod_mQres =[]
    grivan_Qmres =[]
    unionres =[]
    res=[]
    for each in fileset1:
        if(each==final0):
            res = final0res
        elif(each==final1):
            res = final1res
        elif(each==mixmod_mQ):
            res = mixmod_mQres
        elif(each==grivan_Qm):
            res = grivan_Qmres
        elif(each==union):
            res = unionres
        else:
            res= alllouvainres
        fp = open(each,'r')
        fp.readline()
        for line in fp:
            line = line.split()
            params = line[0].split('_')
            alpha = float(params[2])
            p = float(params[3])
            mu = float(params[4])
            p1 = float(params[5])
            p2 = float(params[6])
            gt = float(line[1])
            mod = float(line[2])
            nmi =float(line[3])
            newrow = [alpha, p,mu,p1,p2,gt,mod,nmi]
            res.append(newrow)
        fp.close()
        if(each==final0):
            final0res=copy.deepcopy(res)
        elif(each==final1):
            final1res=copy.deepcopy(res)
        elif(each==mixmod_mQ):
            mixmod_mQres =copy.deepcopy(res)
        elif(each==grivan_Qm):
            grivan_Qmres =copy.deepcopy(res)
        elif(each==union):
            unionres = copy.deepcopy(res)
        else:
            alllouvainres=copy.deepcopy(res)
    return final0res,final1res,alllouvainres,mixmod_mQres,grivan_Qmres,unionres
```

```python
def readdata_infocom():
    infocomres =[]
    fp = open(infocom,'r')
    fp.readline()
    for line in fp:
        line = line.split()
        params = line[0].split('/')
        alpha = float(params[2].split('-')[1])
        p = float(params[3].split('-')[1])
        mu = float(params[4].split('-')[1])
        p1 = float(params[5].split('-')[1])
        p2 = float(params[6].split('-')[1])
        #gt = float(line[1])
        mod = float(line[1])
        nmi =float(line[2])
        newrow = [alpha, p,mu,p1,p2,mod,nmi]
        infocomres.append(newrow)
    fp.close()
    
    return infocomres
```

```python
# Louvain_Qm,Metafac,Compmod
final0res,final1res,alllouvainres,mixmod_mQres,grivan_Qmres,unionres = read_data()
 
final0df = pd.DataFrame(final0res, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
final1df = pd.DataFrame(final1res, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
alllouvaindf = pd.DataFrame(alllouvainres, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
mixmod_mQdf = pd.DataFrame(mixmod_mQres, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
grivan_Qmdf = pd.DataFrame(grivan_Qmres, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
uniondf = pd.DataFrame(unionres, columns = ['alpha', 'p','mu','p1','p2','gt','mod','nmi'])
```

```python
def alphaplots(saveas,pvalues,p1=0.8,p2=0.1,mu=0.05):
    for plot in plottypes:
        filename = saveas+plot+"/ForAlpha_"+plot+"Plot_p"
        for p in pvalues:
            filename1 = filename+"_"+str(int(p*10))
            print(filename1)
            df1 = alllouvaindf[(alllouvaindf["p"]==p)&(alllouvaindf["p1"]==p1)&(alllouvaindf["mu"]==mu)&(alllouvaindf["p2"]==p2)]
            df1 = df1.sort_values(by=['alpha'])

            df2 = final0df[(final0df["p"]==p)&(final0df["p1"]==p1)&(final0df["mu"]==mu)&(final0df["p2"]==p2)]
            df2 = df2.sort_values(by=['alpha']) 

            df3 = final1df[(final1df["p"]==p)&(final1df["p1"]==p1)&(final1df["mu"]==mu)&(final1df["p2"]==p2)]
            df3 = df3.sort_values(by=['alpha'])
            
            df4 = infocomdf[(infocomdf["p"]==p)&(infocomdf["p1"]==p1)&(infocomdf["mu"]==mu)&(infocomdf["p2"]==p2)]
            df4 = df4.sort_values(by=['alpha'])
            
            df5 = mixmod_mQdf[(mixmod_mQdf["p"]==p)&(mixmod_mQdf["p1"]==p1)&(mixmod_mQdf["mu"]==mu)&(mixmod_mQdf["p2"]==p2)]
            df5 = df5.sort_values(by=['alpha'])
            
            df6 = grivan_Qmdf[(grivan_Qmdf["p"]==p)&(grivan_Qmdf["p1"]==p1)&(grivan_Qmdf["mu"]==mu)&(grivan_Qmdf["p2"]==p2)]
            df6 = df6.sort_values(by=['alpha'])
            
            df7 = uniondf[(uniondf["p"]==p)&(uniondf["p1"]==p1)&(uniondf["mu"]==mu)&(uniondf["p2"]==p2)]
            df7 = df7.sort_values(by=['alpha'])
            
            plt.plot(df1['alpha'],df1[plot],label="AllLouvain",marker= 'o')
            plt.plot(df2['alpha'],df2[plot],label = "Final0",marker= 'v')
            plt.plot(df3['alpha'],df3[plot],label = "Final1",marker= 'x')
            #plt.plot(df4['alpha'],df4[plot],label = "InfoCom",marker= '4')
            #plt.plot(df5['alpha'],df5[plot],label = "Mixmod_mQ",marker = 'X')
            plt.plot(df6['alpha'],df6[plot],label = "grivan_Qm",marker = '^')
            plt.plot(df7['alpha'],df7[plot],label = "Union" , marker = '1')
            if plot == 'mod':
                plt.plot(df3['alpha'],df3['gt'],label = "Ground Truth",marker= '.')
            titlesubpart = "plot P = "+str(p)+", p1= "+str(p1)+", p2= "+str(p2)+", mu= "+str(mu)
            plt.title(plot+titlesubpart)
            plt.xlabel('Alpha')
            plt.ylabel(plot)
            plt.legend()
            #df1.plot.line(x='alpha', y='mod')
            #df = alllouvaindf[(alllouvaindf["p"]==0.8)&(alllouvaindf["p1"]==0.8)&(alllouvaindf["mu"]==0.05)&(alllouvaindf["p2"]==0.1)]
            #df  =df.sort_values(by=['alpha'])
            #df.plot.line(x='alpha', y='mod')
            plt.savefig(filename1)
            #plt.show()
            plt.clf()
            
```
