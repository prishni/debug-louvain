```python
def plotalltogether(algoname,resultsfiledir,resultsfilename,pathtosave,comparewithfile,compare,lowerlim, upperlim,label,saveas): 
    gt_mod = []
    mixmod_mod = []
    alllouvain = []
    titlestring = algoname
    axeslabels = []
    x = [] 
    series = []
    labels = []
    i=1
    for a in alpha:
        #print(a)
        netlist=[]
        for p in pval:
            strn='./Raphael_27.6.17/net_networks/network_' + str(a) + '_' + str(p) + '_0.05_0.3_0.0:'
            netlist.append(strn)
            resultsmixmod = open(resultsfiledir+resultsfilename)
            resultsmixmod.readline()
            for line in resultsmixmod:
                line = line.split()

                if(line[0] == strn):
                    #print("line found")
                    #print("Line found alpha: {0} p: {1}".format(a, p))
                    gt_mod.append(float(line[1]))
                    mixmod_mod.append(float(line[2]))
                    axestick = str(a)+"/"+str(p)
                    axeslabels.append(axestick)
                    x.append(i)
                    i+=1
                    #xval.append(p)
                    break
            if(compare==1):
                resultsalllouvain = open(comparewithfile)
                resultsalllouvain.readline()
                for line in resultsalllouvain:
                    line = line.split()

                    if(line[0] == strn):
                        alllouvain.append(float(line[2]))
                        #xval.append(p)
                        break
                resultsalllouvain.close()

            resultsmixmod.close()
            
    
    plt.figure(figsize = (10,10))
    plt.plot(x,gt_mod, label = 'Ground Truth', marker = 'o')
    plt.plot(x,mixmod_mod, label = label, marker = '.')
    if(compare==1):
        plt.plot(x,alllouvain,label = 'All_louvain')
    plt.legend(loc = 'lower right')
    titlestring = titlestring
    plt.title(titlestring,fontsize = fsize)
    plt.ylim([lowerlim, upperlim])
    plt.xlabel("Alpha/P values",fontsize = fsize)
    plt.ylabel("Modularity",fontsize = fsize)
    plt.xticks(x,axeslabels, rotation = 90)
    plt.savefig(pathtosave+saveas+'.png')
    plt.close();
    
```

```python
def _plotalltogether(x, series, labels, saveas, titlestring, xaxeslabels, yaxeslabels=None, xticks = None, yticks = None):
    lowerlim, upperlim = (0.3, 0.7)
    plt.figure(figsize = (12,10))
    markers = ['^', '.', 'x', 'v']
    linestyles = [':', '-.', '--']
    i = 0
    for val, label in zip(series, labels):
        print("Plotting: {0}".format(label))
        plt.plot(x, val, label = label, marker = markers[i%4], linestyle = linestyles[i%3], alpha = 0.7)
        i += 1
    plt.legend(loc = 'lower right')
    plt.title(titlestring,fontsize = fsize)
    plt.ylim([lowerlim, upperlim])
    plt.xlabel("Alpha/P values",fontsize = fsize)
    plt.ylabel("Modularity",fontsize = fsize)
    #print("x: ", x, "axeslabels: ",axeslabels)
    plt.xticks(x,xaxeslabels, rotation = 90)
    #plt.show()
    plt.savefig(saveas+'.png')
    plt.close();
    pass

def plotalltogetherv2(algos,resultsfilenames,saveas,titlestring, compare=0,comparewithfile=None):
    series = []
    labels = []
    axeslabels = []
    for r in range(len(algos)):
        gt_mod = []
        mixmod_mod = []
        alllouvain = []
        x = []
        i=1
        for a in alpha:
            netlist=[]
            for p in pval:
                #strn='./Raphael_27.6.17/net_networks/network_' + str(a) + '_' + str(p) + '_0.05_0.3_0.0:'
                strn= networkpath +'network_' + str(a) + '_' + str(p) + filenamesuffix
                netlist.append(strn)
                resultsmixmod = open(resultsfilenames[r])
                resultsmixmod.readline()
                for line in resultsmixmod:
                    line = line.split()
                    if(line[0] == strn):
                        #print("line found")
                        #print("Line found alpha: {0} p: {1}".format(a, p))
                        gt_mod.append(float(line[1]))
                        mixmod_mod.append(float(line[2]))
                        axestick = str(a)+"/"+str(p)
                        axeslabels.append(axestick)
                        x.append(i)
                        i+=1
                        #xval.append(p)
                        break
                if(compare==1):
                    resultsalllouvain = open(comparewithfile)
                    resultsalllouvain.readline()
                    for line in resultsalllouvain:
                        line = line.split()

                        if(line[0] == strn):
                            alllouvain.append(float(line[2]))
                            #xval.append(p)
                            break
                    resultsalllouvain.close()
                resultsmixmod.close()
        
        series.append(mixmod_mod)
        labels.append(algos[r])
        #def _plotalltogether(x, series, labels, saveas, titlestring, xaxeslabels, yaxeslabels=None, xticks = None, yticks = None):
    if(compare):
        series.append(alllouvain)
        labels.append("All Louvain")
    series.append(gt_mod)
    labels.append("Ground Truth")
    _plotalltogether(x, series, labels, saveas, titlestring, axeslabels)
```
